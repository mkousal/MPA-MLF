# Exercise 1.5-1:

The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The next number in this series above is 8 + 13 = 21



Create a function that takes an integer as an input and return a list that contains all fibonaci numbers smaller than input integer

example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonaci(upper_threshold: int) -> list:
  numbers = [0,1]
  while numbers[-1] + numbers[-2] < upper_threshold:
    numbers.append(numbers[-1] + numbers[-2])
  return numbers

fibonaci(356)
```

```python
def display_as_digi(number: int) -> None:
  nums = [int(x) for x in str(number)]
  print(nums)
  chars = {
    0: "XXXX XX XX XXXX",
    1: "  X  X  X  X  X", 
    2: "XXX  XXXXX  XXX", 
    3: "XXX  XXXX  XXXX",
    4: "X  X  XXX  X  X",
    5: "XXXX  XXX  XXXX",
    6: "XXXX  XXXX XXXX",
    7: "XXX  X  X  X  X",
    8: "XXXX XXXXX XXXX",
    9: "XXXX XXXX  XXXX",
  }
  for y in range(0,5,1):
    for x in range(len(nums)):
      print(chars[nums[x]][y*3:y*3+3:1], end=" ")
    print(" ")
  pass

display_as_digi(1234567890)
```

### 1


- Create a numpy array with shape 5x5, that contains number from 25..to 0
- create a function that takes an arbitary numpy array as an input and sets all numbers smaller than user defined threshold to 0
- test your function on your array
- make one implementation using loops and one withnout using loops, compare execution time with *time* library

```python
import numpy as np
import time

def null_array_items(array, threshold):
    array[np.where(array < threshold)] = 0
    return array

def null_array_items_loop(array, threshold):
    for i, x in np.ndenumerate(array):
        if x < threshold:
            array[i] = 0

    return array

arr = np.arange(25).reshape(5,5)
start_time_no_loop = time.perf_counter()
null_array_items(arr, 15)
print("Time elapsed using no-loop variant\n--- %s sec ---" % (time.perf_counter() - start_time_no_loop))
print(arr)

arr = np.arange(25).reshape(5,5)
start_time_loop = time.perf_counter()
null_array_items_loop(arr, 15)
print("Time elapsed using loop variant\n--- %s sec ---" % (time.perf_counter() - start_time_loop))
print(arr)
```
